```python
import nltk
from nltk.corpus.reader import CHILDESCorpusReader
import numpy as np
import pandas as pd
from nltk.probability import FreqDist
corpus = nltk.data.find('/Users/Ershui13/nltk_data/corpora/childescorpus/')
Eve = CHILDESCorpusReader(corpus, 'Brown/Eve/.*.xml')
```

```python
words = Eve.words(speaker = 'MOT')
tag_words = Eve.tagged_words(speaker = 'MOT')
verbs = [pair for pair in tag_words if pair[-1] == 'v']
```

```python
fdEve = FreqDist(words)
fvEve = FreqDist(verbs)
zipfEve = fdEve.most_common()
zipfEverb = fvEve.most_common()
```

```python
from pandas import DataFrame  
dfwords = DataFrame(zipfEve)
dfverbs = DataFrame(zipfEverb)
dfwords.columns = ['word','count']
dfverbs.columns = ['word','count']
dfwords['rank']=list(range(1,len(zipfEve)+1))
dfverbs['rank']=list(range(1,len(zipfEverb)+1))
dfwords.to_csv('/Users/Ershui13/Desktop/ABSTRACT/Yang/Eve_all_words.csv',index=False)
dfverbs.to_csv('/Users/Ershui13/Desktop/ABSTRACT/Yang/Eve_all_verbs.csv',index=False)
```
