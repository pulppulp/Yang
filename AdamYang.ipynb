```python
import nltk
from nltk.corpus.reader import CHILDESCorpusReader
import numpy as np
import pandas as pd
from nltk.probability import FreqDist
corpus = nltk.data.find('/Users/Ershui13/nltk_data/corpora/childescorpus/')
Adam = CHILDESCorpusReader(corpus, 'Brown/Adam/.*.xml')
```

```python
words = Adam.words(speaker = 'MOT')
tag_words = Adam.tagged_words(speaker = 'MOT')
verbs = [pair for pair in tag_words if pair[-1] == 'v']
```

```python
fdAdam = FreqDist(words)
```

```python
fvAdam = FreqDist(verbs)
```

```python
zipfAdam = fdAdam.most_common()
zipfAverb = fvAdam.most_common()
```

```python
from pandas import DataFrame  
dfwords = DataFrame(zipfAdam)
dfverbs = DataFrame(zipfAverb)
dfwords.columns = ['word','count']
dfverbs.columns = ['word','count']
dfwords['rank']=list(range(1,len(zipfAdam)+1))
dfverbs['rank']=list(range(1,len(zipfAverb)+1))
dfwords.to_csv('/Users/Ershui13/Desktop/ABSTRACT/Yang/Adam_all_words.csv',index=False)
dfverbs.to_csv('/Users/Ershui13/Desktop/ABSTRACT/Yang/Adam_all_verbs.csv',index=False)
```
